```python
# text detection
import pytesseract
from PIL import Image
from easyocr import Reader
import boto3

access_key = ''
secret_access_key = ''
textract_client = boto3.client('textract',
                               aws_access_key_id = access_key,
                               region_name='')

reader = Reader(['en'])

def read_text_tesseract(image_path):

  image_path = '/content/txt_data/journal.jpg'
  text = pytesseract.image_to_string(Image.open(image_path),lang='eng')

  return text

def read_text_easyocr(image_path):
  # Use the correct method to load the image from a path
  image = Image.open(image_path)

  # Use EasyOCR's readtext function to detect text in the image
  results = reader.readtext(image)
  # Initialize an empty string to store the extracted text
  text = ""

  # Loop through the results and extract only the text part
  for result in results:
      text += result[1] + ' '  # result[1] contains the text

  # Remove the trailing space
  text = text.strip()

  # Print the extracted text
  print("Extracted text:")
  return text


def read_text_textract(image_path):
  with open(image_path,'rb') as im:
    response = textract_client.detect_document_text(Document={Bytes:im.read()})
    text = ''
    for item in response {'Blocks'}:
      if item['BlockType'] == 'LINE':
        text = text + item['text']
  text = text[:-1]
  return text
```

```python
# compare performance
import os

def jaccard_similarity(sentence1, sentence2):
    # Tokenize sentences by splitting them into words
    words1 = set(sentence1.lower().split())
    words2 = set(sentence2.lower().split())

    # Calculate the intersection and union of the sets
    intersection = words1.intersection(words2)
    union = words1.union(words2)

    # Compute the Jaccard similarity
    similarity = len(intersection) / len(union) if len(union) != 0 else 0
    return similarity

score_tesseract = 0
score_easyocr = 0
score_textract = 0

for image_path in os.listdir('/content/txt_data'):
    image_path = os.path.join('/content/txt_data', image_path)
    gt = image_path[:-4].replace('_',' ').lower()

    score_tesseract += jaccard_similarity(gt, read_text_tesseract(image_path).lower().replace('\n','').replace('!','').replace('?',''))
    score_easyocr += jaccard_similarity(gt, read_text_easyocr(image_path).lower().replace('\n','').replace('!','').replace('?',''))
    score_textract += jaccard_similarity(gt, read_text_textract(image_path).lower().replace('\n','').replace('!','').replace('?',''))

print('score_tesseract', score_tesseract/20)
print('score_easyocr', score_easyocr/20)
print('score_textract', score_textract/20)
```
